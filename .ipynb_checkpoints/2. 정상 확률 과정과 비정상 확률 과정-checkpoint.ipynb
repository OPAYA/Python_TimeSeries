## 정상 확률 과정

**협의의 정상 확률 과정(strictly stationay process,strong stationary preocess)** 은 확률 과정의 모든 모멘트가 시간차이에만 의존하고 절대 시간에 의존하지 않는 것이다.

이를 수학적으로 표현하면 임의의 $t, s, k_i$에 대해

$$ E[Y_t, Y_{t+k}, Y_{t+k_2}.. .Y_{t+k_i}]= E[Y_s, Y_{s+k_1}, Y_{s+k_2}...Y_{s+k_i}] $$

가 성립한다.

즉 기댓값의 경우

$$E[Y_t] = E[Y_s]$$

가 성립하고

auto-covariance의 경우 

$$E[Y_tY_{t+k}] = E[Y_sY_{s+k}] = f(k)$$

가 성립한다.

위 두가지 조건만 성립하는 경우에는 **weak stationay process** 라고 한다.

stationay process에서는 auto-covariance의 값이 시간 변수의 차이 즉 시차(lag) $k$에만 의존한다. 따라서 자기공분산 시차에 대한 함수이다. 이를 auto covariance function이라고 한다.

$$\gamma_{t, t+k} = \gamma_{0, k}$$

stationay process 과정의 auto-correlation도 마찬가지로 시차 $k$ 에만 의존한다. 이를 auto-correlation function ACF라고 한다.

$$\rho_{t, t+k} = \rho_{0, k}$$

stationary process는 다음과 같은 성질을 만족한다.

$$\rho_0 = Var[Y_t]$$
$$\rho_k = \rho_{-k}$$
$$|\rho_k| \leq \rho_0$$
$$\rho_0 = 1$$
$$\rho_k = \rho_{-k}$$
$$|\rho_k| \leq 1$$

## 에르고딕 성질

정상 확률 과정에서는 각각의 시간에 해당하는 확률 변수의 무조건부 분포가 같다. 따라서 시계열 데이터를 이루는 각 숫자가 하나의 분포에서 나온 표본 데이터 집합이라고 생각할 수 있다. 이 성질을 이용하면 기댓값이나 auto covariance등에서 필요한 앙상블 평균을 계산할 때 여러개의 시계열 데이터 표본이 필욯자ㅣ 않고 단 하나의 시계열 데이터 표본만 있어도 된다.

기댓값의 경우
$$\lim_{x\to\infty} {1\over N}\sum Y_t = E[Y_t]$$

auto-covariance의 경우

$$\lim_{x\to\infty} {1\over N} \sum Y_tY_{t+k}= E[Y_tY_{t+k}]$$

가 성립한다.
이를 에르고딕 성질(ergodicity)이라고 한다.

## 비정상 확률 과정

정살 확률 과정이 아닌 확률 과정이 **비정상 확률 과정(non-stationary process)이다.**
비정상 확률 과정이 되는 경우는
- 추세를 가지는 경우. 일차 모멘트 즉 기댓값 $E[y_t$]이 0이 아니며 시간에 따라 변화함
- 추세가 없지만 $E[y_t] = 0$ 분산 $Var[y_t]$이 시간에 따라 변하는 경우

```python
import matplotlib.pyplot as plt
import statsmodels as sm
import numpy as np
import scipy as sp
import scipy.stats
```

```python
df = sm.datasets.get_rdataset("CanPop", package='carData').data
```

```python
df.plot(x='year', y='population')
plt.xlabel('year')
plt.ylabel('population')
plt.title('Change population through time')
plt.show()
```

다음 시계열 자료들은 동일한 확률 과정의 샘플들이다. 하나 하나의 샘플만 보면 마치 추세가 있는 것처럼 보인다. 그러나 이는 확률 과정의 분산$Var[y_t]$이 시간 $t$에 따라 커지기 때문이다. 그래프에 표시하였듯이 t=400에서의 분산 $Var[Y_{400}]$은 t=100에서의 분산 $Var[y_{100}]$보다 크다. 이런 경우 **stochastic trend** 를 가진다고 말하기도 한다.

```python
N = 500
t1 = 100
t2 = 400
t = np.arange(N)

np.random.seed(12)
y1 = np.insert(np.cumsum(sp.stats.norm.rvs(size=N-1)), 0, 0)
np.random.seed(18)
y2 = np.insert(np.cumsum(sp.stats.norm.rvs(size=N-1)), 0, 0)
np.random.seed(22)
y3 = np.insert(np.cumsum(sp.stats.norm.rvs(size=N-1)), 0, 0)
np.random.seed(24)
y4 = np.insert(np.cumsum(sp.stats.norm.rvs(size=N-1)), 0, 0)

plt.subplot(211)
plt.plot(t, y1)
plt.plot(t, y2)
plt.plot(t, y3)
plt.plot(t, y4)

plt.plot(t1, y1[t1], 'o', markersize=5)
plt.plot(t2, y1[t2], 'o', markersize=5)
plt.plot(t1, y2[t1], 'o', markersize=5)
plt.plot(t2, y2[t2], 'o', markersize=5)
plt.plot(t1, y3[t1], 'o', markersize=5)
plt.plot(t2, y3[t2], 'o', markersize=5)
plt.plot(t1, y4[t1], 'o', markersize=5)
plt.plot(t2, y4[t2], 'o', markersize=5)

plt.subplot(212)
plt.grid(False)
plt.plot(t, sp.stats.norm(t1, 0.08*t1).pdf(t))
plt.plot(t, sp.stats.norm(t2, 0.08*t2).pdf(t))
```
